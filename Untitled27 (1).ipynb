```python
# 1. IMPORTING LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report
```

```python
# 2. LOADING THE DATASET
df = pd.read_csv(r"C:\Users\sumat\Downloads\air quality index.csv")
print("\nDATASET LOADED SUCCESSFULLY\n")
print(df.head())
```

```python
# 3. DATA EXPLORATION
print("\nHEAD:\n", df.head())
print("\nTAIL:\n", df.tail())
print("\nSAMPLE:\n", df.sample(5))

print("\nUNIQUE VALUES:\n", df.nunique())
print("\nDUPLICATES:", df.duplicated().sum())

print("\nINFO:")
df.info()

print("\nDESCRIBE:\n", df.describe(include='all'))

print("\nSHAPE:", df.shape)
print("\nCOLUMNS:", df.columns)
```

```python
# 4. DATA CLEANING (AUTO NUMERIC SELECTION)
# Automatically keep only numeric columns
df = df.select_dtypes(include=['number'])

print("\nNUMERIC COLUMNS USED FOR ML:\n", df.columns)

# Correlation (Safe – numeric only)
print("\nCORRELATION MATRIX:\n", df.corr())

# Missing values
print("\nMISSING VALUES:\n", df.isnull().sum())
df = df.dropna()

# Outlier visualization (first numeric column)
plt.boxplot(df.iloc[:, 0])
plt.title("Outlier Detection")
plt.show()
```

```python
# 5. DATA SCALING

# Last column as y, remaining as X
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
# 6. ENCODING FOR CLASSIFICATION
# (Convert continuous y into 4 classes)
y_class = pd.qcut(y, q=4, labels=[0, 1, 2, 3])
```

```python
# 7. TRAIN–TEST SPLIT

# Regression split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Classification split
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(
    X_scaled, y_class, test_size=0.2, random_state=42
)
```

```python
# 8. BUILD THE MODELS

# Linear Regression
reg_model = LinearRegression()
reg_model.fit(X_train, y_train)

# Logistic Regression
log_model = LogisticRegression(max_iter=5000)
log_model.fit(X_train_c, y_train_c)

# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_c, y_train_c)
```

```python
# 9. TEST THE MODELS

y_pred_reg = reg_model.predict(X_test)
y_pred_log = log_model.predict(X_test_c)
y_pred_rf  = rf_model.predict(X_test_c)
```

```python
# 10. COMPARE ACTUAL vs PREDICTED

comparison = pd.DataFrame({
    'Actual_Y': y_test,
    'Predicted_Y': y_pred_reg
})

print("\nACTUAL vs PREDICTED (REGRESSION):\n")
print(comparison.head(10))
```

```python
# 11. PERFORMANCE EVALUATION

print("\n--- LINEAR REGRESSION ---")
print("MSE:", mean_squared_error(y_test, y_pred_reg))
print("R2 Score:", r2_score(y_test, y_pred_reg))

print("\n--- LOGISTIC REGRESSION ---")
print("Accuracy:", accuracy_score(y_test_c, y_pred_log))
print(classification_report(y_test_c, y_pred_log))

print("\n--- RANDOM FOREST ---")
print("Accuracy:", accuracy_score(y_test_c, y_pred_rf))
print(classification_report(y_test_c, y_pred_rf))
```

```python
# 12. PREDICTION FOR NEW VALUES (NO WARNING)

# Use a sample with proper feature names
new_data = X.iloc[[0]]   # taking first row as example
new_data_scaled = sc.transform(new_data)

new_pred_reg = reg_model.predict(new_data_scaled)
new_pred_log = log_model.predict(new_data_scaled)
new_pred_rf  = rf_model.predict(new_data_scaled)

print("\nNEW DATA PREDICTION")
print("Predicted Value (Linear Regression):", new_pred_reg)
print("Predicted Class (Logistic Regression):", new_pred_log)
print("Predicted Class (Random Forest):", new_pred_rf)
```
